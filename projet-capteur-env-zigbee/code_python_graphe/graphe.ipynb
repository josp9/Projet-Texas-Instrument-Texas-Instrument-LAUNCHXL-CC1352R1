```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(file):
    temperature = []
    humidity = []

    with open(file, 'r') as f:
        for line in f:
            data = line.strip().split(',')

            
            if len(data) >= 2:
                temperature.append(float(data[0]))
                humidity.append(float(data[1]))
                

    return np.array(temperature), np.array(humidity)

def graphic_draw(temperature, humidity):
 
    plt.figure(figsize=(6, 4))
    
    plt.plot(temperature, label='Temp (°C)', marker='o')

    
    plt.plot(humidity, label='Humid (%)', marker='o')
    
    
    plt.title('Graphique de Temp et Humid')
    plt.xlabel('Échantillons')
    plt.ylabel('Valeurs')
    plt.legend()
    

    plt.grid(True)
    plt.show()

file_data = 'capteur_sht.txt'


temperature, humidity= read_data(file_data )


graphic_draw(temperature, humidity)
```

```python
def read_data_gaz(file):
    data_GM_102B = []
    data_GM_302B = []
    data_GM_502B = []
    data_GM_702B = []

    with open(file, 'r') as f:
        for line in f:
            data = line.strip().split(',')

            
            if len(data) >= 4:
                data_GM_102B.append(int(data[0]))
                data_GM_302B.append(int(data[1]))
                data_GM_502B.append(int(data[2]))
                data_GM_702B.append(int(data[3]))
                

    return np.array(data_GM_102B), np.array(data_GM_302B), np.array(data_GM_502B), np.array(data_GM_702B)

def graphic_draw_gaz(data_GM_102B, data_GM_302B, data_GM_502B, data_GM_702B):
 
    plt.figure(figsize=(6, 4))
    
    plt.plot(data_GM_102B, label='GM_102B', marker='o')
    
    plt.plot(data_GM_302B, label='GM_302B', marker='o')
    
    plt.plot(data_GM_502B, label='GM_502B', marker='o')
    
    plt.plot(data_GM_702B, label='GM_702B', marker='o')
    
    
    
    plt.title('Graphique de GM_102B, GM_302B, GM_502B, GM_702B')
    plt.xlabel('Échantillons')
    plt.ylabel('Valeurs')
    plt.legend()
    

    plt.grid(True)
    plt.show()

file_data = 'gaz.txt'


data_GM_102B, data_GM_302B, data_GM_502B, data_GM_702B = read_data_gaz(file_data )


graphic_draw_gaz(data_GM_102B, data_GM_302B, data_GM_502B, data_GM_702B)
```
